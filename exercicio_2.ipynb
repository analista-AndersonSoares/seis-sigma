```python
# importando as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# carregando o arquivo Excel e exibindo o DataFrame
df = pd.read_excel("[FM2S]Lista_Exercícios_1.xlsx", sheet_name='02.Clínica')
print(df.head(10))
```

```python
# Criação de variáveis para cada coluna do DataFrame
col_paciente = df['Paciente'] 
col_dia = df['Dia']
col_especialidade = df['Especialidade']
col_satisfacao = df['Satisfação']

print(col_dia)
```

```python
print("Letra A: O índice de satisfação (nota de 0 a 10) é tratado na estatística como uma variável contínua")
```

```python
df_media_dia = (
    df.groupby("Dia")["Satisfação"]
      .mean()
      .reset_index()
)
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    data=df,
    x="Satisfação",
    bins=10,
    kde=True,
    color="#457b9d"
)

plt.title("Letra B - Distribuição Geral da Satisfação")
plt.xlim(0, 10)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x="Especialidade",
    y="Satisfação",
    palette="Set2"
)

plt.title("Letra C - Satisfação por Especialidade")
plt.ylim(0,10)

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.scatterplot(
    data=df,
    x=df.index,
    y="Satisfação",
    hue="Especialidade",
    palette="Set2",
    alpha=0.7
)

plt.axvline(100, color="red", linestyle="--", label="Fim do 10º Dia")
plt.title("Letra E/F - Tendência das Notas")
plt.xlabel("Ordem de Atendimento")
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.lineplot(
    data=df_media_dia,
    x="Dia",
    y="Satisfação",
    marker="o",
    linewidth=2
)

plt.axvline(10, color="red", linestyle="--", label="10º Dia")
plt.title("Letra H/I - Média da Satisfação por Dia")
plt.xticks(np.arange(1,21))
plt.ylim(0,10)
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(6,5))

ax = sns.countplot(
    data=df,
    x="Especialidade",
    palette="Set2"
)

ax.bar_label(ax.containers[0])

plt.title("Letra J - Quantidade de Pacientes por Especialidade")
plt.ylabel("Quantidade")

plt.show()
```
